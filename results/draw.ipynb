```python
import os

# 定义一个函数来提取数字
def extract_numbers(line):
    # 分割行并获取第四个元素
    parts = line.split()
    if len(parts) >= 5:
        return float(parts[4])
    else:
        return None

# 获取所有txt文件
txt_files = [file for file in os.listdir('.') if file.endswith('.txt')]

# 创建一个字典来存储数字数组，其中键是文件名（去除扩展名），值是数字数组
data_dict = {}

# 遍历每个txt文件并提取数字
for txt_file in txt_files:
    with open(txt_file, 'r') as file:
        lines = file.readlines()
        found_loaded_gt = False
        numbers = []
        for line in lines:
            if found_loaded_gt:
                # 提取数字
                number = extract_numbers(line)
                if number is not None:
                    numbers.append(number)
            elif line.startswith("Loaded gt"):
                found_loaded_gt = True
        # 去除文件扩展名作为键，将数字数组存储到字典中
        file_name_without_extension = os.path.splitext(txt_file)[0]
        data_dict[file_name_without_extension] = numbers

# 打印每个文件名和对应的数字数组
# for file_name, numbers in data_dict.items():
#     print(f"{file_name}: {numbers}")
```

```python
# 创建一个新的字典，用于存储除法后的数据
new_data_dict = {}

# 指定multicand_1的名称
multicand_1_name = "multicand_1"
multicand_1_data = data_dict[multicand_1_name]

# 遍历data_dict中的每个项
for label, data in data_dict.items():
    # 将当前数组的每个数字除以multicand_1对应位置的数字
    new_data = [x / y for x, y in zip(data, multicand_1_data)]
    # 将新的数据添加到新的字典中
    new_data_dict[label] = new_data

# 打印新的字典
for label, data in new_data_dict.items():
    print(f"{label}: {data}")
```

```python
import matplotlib.pyplot as plt

x_coords = []

# 生成[1, 30)的整数
for i in range(1, 30):
    x_coords.append(i)

# 生成[30, 100)的10的倍数
for i in range(30, 100, 10):
    x_coords.append(i)

# 生成[100, 500)每次增加40的数字
for i in range(100, 500, 40):
    x_coords.append(i)
```

```python
# 绘制折线图
for label, data in new_data_dict.items():
    plt.plot(x_coords, data, label=label, marker='o', markersize=3.5)

plt.axhline(y=1.2, color = "red", linestyle='--')

plt.legend()
plt.xlabel("ef")
```
